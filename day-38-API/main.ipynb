```python
import requests
import json

headers_nutri = {
    'x-app-id': NUTRI_APPLICATION_ID,
    'x-app-key': NUTRI_APPLICATION_KEY,
    'Content-Type': 'application/json'
}
```

```python
nutrition_text = input("Enter the food you ate: ")

def get_nutritionix_data(query, headers):

    data = {
        'query': query,
    }
    
    response = requests.post(NUTRI_NUTRITION_ENDPOINT, headers=headers, data=json.dumps(data))
    return response.json()

get_nutritionix_data(nutrition_text, headers_nutri)
```

```python
exercise_text = input("Enter the excercise you did: ")

def get_exercise_data(query, headers):

    parameters = {
        'query': query,
        'gender': GENDER,
        'weight_kg': WEIGHT_KG,
        'height_cm': HEIGHT_CM,
        'age': AGE,
    }

    response = requests.post(NUTRI_EXCERCISE_ENDPOINT, headers=headers, json=parameters)
    
    return response.json()

exercise_data = get_exercise_data(exercise_text, headers_nutri)
exercise_data
```

```python
headers_sheety = {
    'Authorization': f'BASIC {SHEETY_APPLICATION_KEY}',
}
```

```python
def add_data_to_sheety(headers, exercise_data):
    response = requests.post(SHEETY_ENDPOINT, headers=headers, json=exercise_data)
    return response.json()
```

```python
from datetime import datetime

def parse_data_for_sheety(headers_sheety, exercise_data):
    for exercise in exercise_data['exercises']:
        workout = {
            "workout": {
                "date": datetime.now().strftime("%d/%m/%Y"),
                "time": datetime.now().strftime("%H:%M:%S"),
                "exercise": exercise['name'].title(),
                "duration":exercise['duration_min'],
                "calories": exercise['nf_calories'],
            }
        }
    
        sheet_response = add_data_to_sheety(headers_sheety, workout)
        print(sheet_response)

parse_data_for_sheety(headers_sheety, exercise_data)
```

```python
def get_data_from_sheety(headers):
    response = requests.get(SHEETY_ENDPOINT, headers=headers)
    return response.json()

data = get_data_from_sheety(headers_sheety)
data
```
