```python
# Get yesterday's and the day before yesterday's closing price
def get_stock_data():
    
    params = {
    "function": "TIME_SERIES_DAILY",
    "symbol": STOCK,
    "apikey": STOCK_API_KEY
    }

    response = requests.get(AV_ENDPOINT, params=params)
    response.raise_for_status()
    stock_data = response.json()["Time Series (Daily)"]
    return stock_data
    
```

```python
def calculate_delta(stock_data: dict) -> float:  
    data = [value for (key, value) in stock_data.items()]
    yesterday_close = float(data[0]["4. close"])
    day_before_yesterday_close = float(data[1]["4. close"])
    delta_percentage = (yesterday_close/day_before_yesterday_close-1)*100
    return delta_percentage
```

```python
def get_news(): 
    params = {
        "q": COMPANY_NAME,
        "apiKey": NEWS_API_KEY,
    }

    response = requests.get("https://newsapi.org/v2/everything", params=params)
    response.raise_for_status()
    news_data = response.json()["articles"]
    return news_data
    
```

```python
def prepare_news_message(news_data: list, delta_percentage: float) -> str:
    news_message = f"{STOCK}: {round(delta_percentage)}%\n"
    for i in range(3):
        news_message += f"Headline: {news_data[i]['title']}\n"
        news_message += f"Brief: {news_data[i]['description']}\n"
    return news_message
```

```python
news_message = prepare_news_message(get_news(), calculate_delta(get_stock_data()))
news_message
```

```python
import time 
import pywhatkit
import pyautogui
from pynput.keyboard import Key, Controller

def send_whatsapp_message():
    try:
        keyboard = Controller()
        news_message = prepare_news_message(get_news(), calculate_delta(get_stock_data()))
        pywhatkit.sendwhatmsg_instantly(
            phone_no=MY_PHONE_NUMBER, 
            message=news_message,
            tab_close=True
        )
        time.sleep(10)
        pyautogui.click()
        time.sleep(2)
        keyboard.press(Key.enter)
        keyboard.release(Key.enter)
        print("Message sent!")
    except Exception as e:
        print(str(e))
```
