```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = 'https://appbrewery.github.io/news.ycombinator.com/'
```

```python
soup = BeautifulSoup(requests.get(URL).text, 'html.parser')
print(soup.prettify())
```

```python
title = soup.title
title.text
```

```python
article_tag = soup.find_all('a', class_='storylink')
article_text = [article.getText() for article in article_tag]
article_link = [article.get('href') for article in article_tag]
article_upvote = [int(score.getText().split(' ')[0]) for score in soup.find_all('span', class_='score')]
```

```python
for i in range(len(article_upvote)):
    if article_upvote[i] == max(article_upvote):
        print(article_text[i], article_link[i])
```
